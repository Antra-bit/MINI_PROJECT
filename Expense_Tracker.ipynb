```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load existing data
try:
    df = pd.read_csv('expenses.csv')
except FileNotFoundError:
    df = pd.DataFrame(columns=["Date", "Category", "Amount", "Description"])
```

```python
date= input("Enter date (YYYY-MM-DD): ") or datetime.today().strftime('%Y-%m-%d')
category = input("Enter category (eg, food , travel, bills): ")
amt = float(input("Enter Amount"))
desc=input("Enter Description of expenses")

new_expense = pd.DataFrame([[date, category, amt, desc]], columns=df.columns)
df = pd.concat([df, new_expense], ignore_index=True)


# Save to CSV
df.to_csv('expenses.csv', index=False)
print("Expense added!")
```
